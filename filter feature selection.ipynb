```python
goods = mydata[mydata['Fraud'] == 0]
goods.shape
```

```python
bads = mydata[mydata['Fraud'] == 1]
bads.shape
```

```python
numgoods = len(goods)
numbads = len(bads)
```

```python
KS_FDR = pd.DataFrame(np.zeros((nvars+2,3)))
i = 0
for column in mydata:
#     print(i, column)
    KS_FDR.loc[i, 0] = column
    i= i+1
```

```python
KS_FDR.columns = ['field', 'ks', 'FDR']
```

```python
%%time
# this is the KS routine in scipy
j = 0
for column in mydata:
    (ks, pval) = sps.ks_2samp(goods[column],bads[column])
    KS_FDR.loc[j, 'ks'] = ks
    j = j+1
```

```python
%%time
# this calculates the FDR @ 3% for all the variables
# note that you have to calculate it two ways since you don't know which way the bads are ranked by each variable
# calculate it both ways, then take the max
topRows = int(round(len(mydata)*0.03))
j = 0
for column in mydata:
    temp1 = mydata.sort_values(column,ascending=False).head(topRows)
    temp2 = mydata.sort_values(column,ascending=True).head(topRows)
    needed1 = temp1.loc[:,'Fraud']
    needed2 = temp2.loc[:,'Fraud']
    FDR1 = sum(needed1)/numbads
    FDR2 = sum(needed2)/numbads
    FDRate = np.maximum(FDR1,FDR2)
#     print(column, FDR1, FDR2, FDRate)
    KS_FDR.loc[j, 'FDR'] = FDRate
    j = j + 1
```

```python
KS_FDR.sort_values(by=['ks'], ascending=False, inplace=True)
KS_FDR.to_csv('Feature selection filter sorted by KS.csv', index = False)
```
